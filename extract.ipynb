```python
## Dependencies

import requests
import json
import prettytable
import os
import csv
import pymongo
```

```python
csvpath = os.path.join("./",'nat_sector_codes.csv')
csvnatdty = os.path.join("./",'nat_dataty_codes.csv')
print(csvpath)

nat_sector_codes =[]
nat_dataty_codes =[]
```

```python
with open(csvpath, newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    print(csvreader)
    
    for row in csvreader:
        nat_sector_codes.append(row)
```

```python
with open(csvnatdty, newline='') as csvnatdty:
    nat_dty = csv.reader(csvnatdty, delimiter=",")
    print(nat_dty)
    
    for row in nat_dty:
        nat_dataty_codes.append(row)
    
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.nat_earn_ind
collection = db.Worker
```

```python
#Create hours, overtime hours, and average earnings.
pre = 'CES'

ind = '000000' # All Industries in sector

data_ty_ee = '01'

for x in nat_sector_codes:
    series_id = (f"{pre}{x[0]}{ind}{data_ty_ee}")
    sec_code = x[0]
    sec_name = x[1]
    headers = {'Content-type': 'application/json'}
    data = json.dumps({"seriesid": [series_id],"startyear":"2011", "endyear":"2014"})
    p = requests.post('https://api.bls.gov/publicAPI/v2/timeseries/data/', data=data, headers=headers)
    json_data = json.loads(p.text)
    for series in json_data['Results']['series']:
        seriesId = series['seriesID']
        for item in series['data']:
            year = item['year']
            period = item['period']
            month = item['periodName']
            value = item['value']
            
            #Dictonary to be inserted in MongoDB
            post = {
                'SectorCode': sec_code,
                'SectorName': sec_name,
                'SeriesID': seriesId,
                'Year': year,
                'Month': month,
                'Period': period,
                'Total EE': value,
            }
            
            # Insert dictionary into MongoDB as a document
            collection.insert_one(post)
```
